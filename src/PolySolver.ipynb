```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_poly():
    x0 = np.random.randint(-10,10)
    x1 = x0 + np.random.randint(1,10)
    x2 = x1 + np.random.randint(1,10)
    k = np.random.random()*np.random.randint(-10,10)
    while (abs(k)<0.01):
        k = np.random.random()*np.random.randint(-10,10)
    a3 = k
    a2 = -k*(x0+x1+x2)
    a1 = k*(x0*x1+x1*x2+x2*x0)
    a0 = -k*x0*x1*x2
    return a3,a2,a1,a0,x2,x1,x0,k
```

```python
def eval_poly(x,a3,a2,a1,a0):
    return a3*x**3 + a2*x**2 + a1*x**1 + a0*x**0
```

```python
a3,a2,a1,a0,x2,x1,x0,k = make_poly()
print(x0,x1,x2)
x = np.linspace(x0-0.5,x2+0.5)
plt.plot(x,eval_poly(x,a3,a2,a1,a0));
```

```python
x = np.linspace(3,11)
plt.plot(x,eval_poly(x,-1.629597,37.470725,-267.253862,573.618046));
```

# Make pool

```python
import pooler as pooler
number_of_questions = 10
my_pool = pooler.setup_pool(number_of_questions)
```

```python
for i in range(number_of_questions):
    question = my_pool.create_question()    
    a3,a2,a1,a0,x2,x1,x0,k = make_poly()
    my_text_poly = """
    <p>Find the zeros (solutions) of the following polynomial using a Newton type algorithm\("""
    my_text_poly += str('{:f}'.format(a3)) + "x^3 + " + str('{:f}'.format(a2)) + "x^2 + " + str('{:f}'.format(a1)) + "x^1 + " + str('{:f}'.format(a0)) + "\)</p>"
    my_text_poly += "<p>What is the value of the second largest solution you obtained?</p>"
        
    my_pool.set_question_text(question, my_text_poly)
    my_pool.set_question_accuracy(question, x1, 0.1)
    my_pool.set_question_answer(question, x1)
    my_pool.set_question(question)        
```

```python
# finally output
my_pool.tree.write('poly.dat')
```
